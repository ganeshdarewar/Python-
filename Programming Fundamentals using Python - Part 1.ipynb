```python
#lex_auth_01269444890062848087
# https://infytq.onwingspan.com/web/en/viewer/hands-on/lex_auth_01269444890062848087_shared?collectionId=lex_auth_0125409616243425281061_shared&collectionType=Course&pathId=lex_auth_012697946426499072230_shared
def find_correct(word_dict):
    correct, almost, wrong = 0, 0, 0
    for key, value in word_dict.items():
        if key == value:
            correct += 1
        elif len(key) != len(value) or sum(key[i] != value[i] for i in range(len(key))) > 2:
            wrong += 1
        else:
            almost += 1
    return [correct, almost, wrong]

word_dict={'RADIO': 'RADICAL', 'MIND': 'MUND', 'CHECK': 'CHEK', 'ALWAYS': 'ALLISWELL', 'VENDOR': 'VENDING'}
print(find_correct(word_dict))
```

```python
def sms_encoding(data):
    word=data.split()
    st=""
    for i in word:
        if(len(i)==1):
            st=st+i
        else:
            for j in i:
                if j not in "aeiouAEIOU":
                    st=st+j
        st=st+" "
    return st[0:-1]
data="Have a nice day"
print(sms_encoding(data))
```

```python
# https://infytq.onwingspan.com/web/en/viewer/hands-on/lex_auth_0127382283825971201450_shared?collectionId=lex_auth_0125409616243425281061_shared&collectionType=Course&pathId=lex_auth_012697946426499072230_shared
def max_frequency_word_counter(data):
    word = ""
    frequency = 0
    words = data.split()
    words = [word.lower() for word in words]
    for current_word in words:
        current_word_count = words.count(current_word)
        if current_word_count > frequency:
            frequency = current_word_count
            word = current_word
        elif current_word_count == frequency:
            if len(current_word) > len(word):
                word = current_word
    print(word, frequency)

data = "Listen to the big clock Tick tock tick"
max_frequency_word_counter(data)
```
